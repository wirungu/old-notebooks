## 1. Convergence Rates

```python
from pylab import *

def f(n):
    return (1+1/n)**n

f_limit = exp(1)

def error(n):
    return abs(f_limit - f(n))

error_vec = vectorize(error)

x = array(1.2**arange(1,40),dtype=int)

figure()
loglog(x,error_vec(x),'*')
loglog(x,1.3/x) # fit a straight line
show()
```

## 2. Samples from the Normal Distribution

```python
from pylab import *

N = 2000
#seed(2) # to initialize the random number generator
Gauss = normal(size=N) # sample from normal distribution

figure()
plot(Gauss,'x')
axhline(mean(Gauss),color='r')
axhline(mean(Gauss)+std(Gauss),color='g')
axhline(mean(Gauss)-std(Gauss),color='g')
show()
# Note: use sort(Gauss) to sort the samples
```
