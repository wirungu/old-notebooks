## 1. Some matrix notation

```python
from pylab import *

S = arange(0,13)

print(ones_like(S)) # creates an array of 1's in the same shape as S
print(ones(len(S))) # does the same thing in this case

# with "newaxis", we can tell python how exactly to deal with different
# arrays, e.g., how exactly "matrix times vector" is to be interpreted:
A1 = ones((6,6))*arange(6)[:,newaxis] 
A2 = ones((6,6))*arange(6)[newaxis,:]
# Another example how to create an upper triangular matrix of "False" values, maybe useful for HW sheet 7, Problem 4
A_tri = arange(6)[:,newaxis]>arange(6)[newaxis,:]

print(A1)
print(A2)
print(A_tri)

# The "roll" command shifts the entries of an array, as in the following examples (maybe useful for HW sheet 7, Problem 4)
B = arange(1,9)
print(B,roll(B,2), roll(A1,3,axis=0))
```

## 2. Color plots of arrays

```python
n = 10

x = arange(0,n+1)
y = arange(0,n+1)

# Create an example
C = x[newaxis,:]**2 + 4*y[:,newaxis]

print(C)

figure()

# A colored plot with different options
imshow(C,cmap=cm.gnuplot,aspect='1',extent=(0,10,20,30),interpolation='gaussian')    
colorbar()
show()

# Log color scheme
imshow(C,norm=matplotlib.colors.LogNorm())
colorbar()
show()
```

## 3. Masked arrays

```python
# Masked arrays, useful for the plotting in HW sheet 7, Problem 4

D = array([[1,2],[3,4]])

print(D)

E = ma.masked_array(D,mask=[[1,0],[0,1]])

print(E)

imshow(D)
show()
imshow(E)
show()

C_mask = ma.masked_array(C,mask=ones((n+1,n+1))-eye(n+1))

imshow(C_mask)
show()
```
